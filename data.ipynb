```python
import requests

def fetch_memecoins_by_volume(
    vs_currency="usd", 
    order="volume_desc", 
    category="memes", 
    per_page=50, 
    page=1
):
    """
    Fetch coins from a given category (e.g. 'memes', 'solana-ecosystem', etc.)
    ordered by volume, from CoinGecko's API.
    """
    url = "https://api.coingecko.com/api/v3/coins/markets"
    params = {
        "vs_currency": vs_currency,
        "order": order,               # volume_desc sorts results by highest volume first
        "category": category,         # e.g., 'memes' or 'solana-ecosystem'
        "per_page": per_page,
        "page": page,
        "sparkline": "false"
    }
    
    response = requests.get(url, params=params)
    data = response.json()
    
    if not isinstance(data, list):
        print("Unexpected response:", data)
        return []
    
    return data

if __name__ == "__main__":
    # Example 1: Fetch top 50 'meme' coins by volume
    memecoins = fetch_memecoins_by_volume(category="memes")
    
    # Print basic info
    for coin in memecoins:
        print(f"{coin['name']} ({coin['symbol'].upper()}) - "
              f"Current Price: ${coin['current_price']} - "
              f"24h Volume: ${coin['total_volume']}")

    # Example 2: Fetch top 50 Solana ecosystem coins by volume
    solana_coins = fetch_memecoins_by_volume(category="solana-ecosystem")
    
    # Print basic info
    print("\nSolana Ecosystem Coins by 24h Volume:")
    for coin in solana_coins:
        print(f"{coin['name']} ({coin['symbol'].upper()}) - "
              f"Price: ${coin['current_price']} - "
              f"24h Volume: ${coin['total_volume']}")
```
